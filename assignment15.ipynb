```python
#Q2-

print('''

        Data Definition Language (DDL) is used to create and modify the structure of objects in a database 
        using predefined commands and a specific syntax.
        
        CREATE TABLE:
                    The CREATE TABLE command creates a new table in the database.
                    
        eg.
            CREATE TABLE Persons (
                PersonID int,
                LastName varchar(255),
                FirstName varchar(255),
                Address varchar(255),
                City varchar(255)
                );
                
           
        ALTER TABLE
            The ALTER TABLE command adds, deletes, or modifies columns in a table.

            The ALTER TABLE command also adds and deletes various constraints in a table.
            
        eg.
            ALTER TABLE Persons ADD Email varchar(255);
            
        
        DROP TABLE
            The DROP TABLE command deletes a table in the database.


        Example:
            DROP TABLE Persons;
        
        
        TRUNCATE TABLE
            The TRUNCATE TABLE command deletes the data inside a table, but not the table itself.

        Example:
            TRUNCATE TABLE Persons;
        ''')
```
